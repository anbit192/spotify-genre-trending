- Get track detail infos from API (RUN ONLY ONCE!!!)

```python
end = vietnamese_data.shape[0]
step = 50
artist_ids = []
artist_names = []

for i in range(0, end, step):
    end = min(i + step, end)
    # print(min(i + 50, 125))
    track_ids = vietnamese_data.iloc[i:end]["spotify_id"].to_numpy()
    res = sp.tracks(tracks=track_ids)

    for track in res["tracks"]:
        artist_infos = track["artists"]
        
        for artist in artist_infos:
            artist_names.append(artist["name"])
            artist_ids.append(artist["id"])
    
```

- Get all artist detail info from API (RUN ONLY ONCE!!!)

```python
end = artist_ids.shape[0]
step = 50
artist_genres = {}

for i in range(0, end, step):
    end_ = min(i + step, end)
    # print(min(i + 50, 125))
    artists_idx = artist_ids[i:end_]
    res_artists = sp.artists(artists_idx)

    for artist in res_artists["artists"]:
        name = artist["name"]
        id = artist["id"]
        genres = artist["genres"]
        artist_genres[name] = genres
```
