```python
end = df.shape[0]
step = 50
artist_ids = []
artist_names = []

for i in range(0, end, step):
    print("Láº§n thá»©", i)
    end_ = min(i + step, end)
    # print(min(i + 50, 125))
    track_ids = df.iloc[i:end_]["spotify_id"].to_numpy()
    print(track_ids)

    try:
        res = sp.tracks(tracks=track_ids)
        time.sleep(1)
    except SpotifyException as e:
        if e.http_status == 404:
            print(f"Error 404. Skipping.\n===========")
            continue
        else:
            print(f"An unexpected error occurred {e}")
            continue

    for track in res["tracks"]:
        if track is None:
            print("Track is None, skipping.")
            continue
        
        artist_infos = track["artists"]
        
        for artist in artist_infos:
            artist_names.append(artist["name"])
            artist_ids.append(artist["id"])
```

```python
artist_ids = np.unique(artist_ids)

end = artist_ids.shape[0]
step = 50
artist_genres = {}
artist_json = {}

for i in range(0, end, step):
    print("Láº§n thá»©", i)
    end_ = min(i + step, end)
    # print(min(i + 50, 125))
    artists_idx = artist_ids[i:end_]
    res_artists = sp.artists(artists_idx)
    time.sleep(1)

    for artist in res_artists["artists"]:
        name = artist["name"]
        id = artist["id"]
        genres = artist["genres"]

        artist_json[id] = dict({"name": name, "genres": genres})
```
