```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file CSV
df = pd.read_csv("../result/top-daily-vn-final.csv")
```

```python
# Lọc các bài nhạc có daily_movement > 0
df_movement_positive = df[df['daily_movement'] > 0]

# Vẽ biểu đồ đếm số lượng bài hát tăng hạng theo từng thể loại
plt.figure(figsize=(10, 6))
sns.countplot(data=df_movement_positive, x='song_genre', palette='viridis', order=df_movement_positive['song_genre'].value_counts().index)
plt.xlabel('Thể loại')
plt.ylabel('Số lượng bài nhạc tăng hạng')
plt.title('Phân phối các bài nhạc có daily_movement > 0 theo thể loại')
plt.xticks(rotation=45)
plt.show()
```

```python
# Lọc các bài nhạc có daily_movement > 0
df_movement_positive = df[df['weekly_movement'] > 0]

# Vẽ biểu đồ đếm số lượng bài hát tăng hạng theo từng thể loại
plt.figure(figsize=(10, 6))
sns.countplot(data=df_movement_positive, x='song_genre', palette='viridis', order=df_movement_positive['song_genre'].value_counts().index)
plt.xlabel('Thể loại')
plt.ylabel('Số lượng bài nhạc tăng hạng')
plt.title('Phân phối các bài nhạc có weekly_movement > 0 theo thể loại')
plt.xticks(rotation=45)
plt.show()
```

```python
# Vẽ biểu đồ KDE cho mỗi đặc điểm
features = ['energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness']
plt.figure(figsize=(15, 12))
for i, feature in enumerate(features, 1):
    plt.subplot(3, 2, i)
    sns.kdeplot(df[feature], shade=True)
    plt.title(f'Distribution of {feature.capitalize()}')
    plt.xlabel(feature.capitalize())
plt.tight_layout()
plt.show()
```
